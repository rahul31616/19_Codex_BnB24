```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = [-3,5,7]
```

```python
y = [10,2,5]
```

```python
fig = plt.figure(figsize = (15,3))

plt.plot(x,y)
plt.xlim(1,10)
plt.ylim(1,8)
plt.xlabel('X Axis')
plt.ylabel('Y axis')
plt.title('Line plot')
plt.suptitle('Sales Comparasion',size =20,y=1.03)
```

```python
fig.savefig('example.png',dpi = 300, bbox_inches = 'tight')
```
