```python
import pandas as pd

# Fixed dictionary syntax errors
data1 = {
    'Name': ['jai', 'Anuj', 'Jai', 'Princi', 'Gaurav', 'Anuj', 'Princi', 'Abhi'],
    'Age': [27, 24, 22, 32, 33, 36, 27, 32],
    'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Aligarh', 'Nagpur', 'Kanpur', 'Allahabad', 'Aligarh'],
    'Qualification': ['Msc', 'MA', 'MCA', 'Phd', 'B.tech', 'B.com', 'Msc', 'MA']
}

df = pd.DataFrame(data1)
print("Original DataFrame")
print(df)

# Grouping the DataFrame by 'Name'
gk = df.groupby('Name')
print("\nFirst entry in each group:")
print(gk.first())
```
